```python
import pandas as pd
import numpy as np
import os
import re

os.chdir('waterloo_datathon_datasets')
pd.set_option('max_columns', 50)
```

```python
df = pd.read_csv('demographics.csv')

print(df.head())
print(df.tail())
```

```python
# format county column

# drop ny state row
df = df[df['geography'] != 'New York']

# remove ' County, New York'
df.loc[:, 'geography'] = df.geography.apply(lambda x: re.sub(' County, New York', '', x.strip()))

df.head()
```

```python
# create % columns

df.loc[:, '%_9999_or_less'] = df['$9,999_or_less'].astype(float) / df['total_households'].astype(float)
df.loc[:, '%_150000_or_more'] = (df['$150,000_to_$199,999'].astype(float) + df['$200,000_or_more'].astype(float)) / df['total_households'].astype(float)
df.loc[:, '%_health_insurance'] = df['pop_w_health_insurance'].astype(float) / df['population'].astype(float)
df.loc[:, '%_private_coverage'] = df['pop_w_private_health'].astype(float) / df['population'].astype(float)
df.loc[:, '%_public_coverage'] = df['pop_w_public_coverage'].astype(float) / df['population'].astype(float)

print(df.columns)

df.head()
```

```python
# drop unnecessary columns
df.drop(['$9,999_or_less',
       '$10,000_to_$14,999', '$15,000_to_$24,999', '$25,000_to_$34,999',
       '$35,000_to_$49,999', '$50,000_to_$74,999', '$75,000_to_$99,999',
       '$100,000_to_$150,000', '$150,000_to_$199,999', '$200,000_or_more', 'pop_w_health_insurance',
       'pop_w_private_health', 'pop_w_public_coverage',
       'pop_no_health_insurance'], axis=1, inplace=True)

df.head()
```

```python
df.to_csv('demographics_cleaned.csv', index=False)
```
